```python
import csv
import pandas as pd
import os
import altair as alt
import altair_viewer
```

```python
alt.renderers.enable("html")
```

```python
df = pd.read_csv("crime_categories.csv")
```

```python
df.head()
```

```python
df = df.melt(id_vars='Unnamed: 0', var_name='Type', value_name='Count')
```

```python
chart = alt.Chart(df).mark_bar().encode(
    x = "Unnamed: 0\:O",
    y = "Count:Q",
    color = "Type:N",
    tooltip = ["Unnamed:\0", "Type", "Count"]
).properties(
    width = 600,
    height = 400,
    title = "Crime Counts at Stanford by Year"
).configure_axisX(
    labelAngle = 0
)
```

```python
altair_viewer.show(chart)
```
